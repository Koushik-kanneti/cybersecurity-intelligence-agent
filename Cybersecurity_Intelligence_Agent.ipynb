```python
!pip install transformers
!pip install torch
```

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
```

```python
def detect_phishing_zeroshot(email_text):
    labels = ["phishing", "safe", "promotional", "spam"]
    result = classifier(email_text, candidate_labels=labels)

    print(f"📧 Email: {email_text}")
    for lbl, score in zip(result["labels"], result["scores"]):
        print(f" - {lbl}: {score:.2f}")

    print(f"🔍 Prediction: {result['labels'][0].upper()}")

# Test safe
detect_phishing_zeroshot("Hi John, I’ve attached the report. Please review it and send your feedback.")

# Test phishing
detect_phishing_zeroshot("Your PayPal account has been restricted. Click here to verify your identity now.")
```

#IOC Log Scanning in Google Colab

```python
sample_logs = """
192.168.1.101 accessed http://example.com
10.0.0.55 connected to 45.83.120.24 on port 443
powershell.exe -nop -w hidden -enc aGVsbG8=
User logged in from IP 8.8.8.8
Suspicious file: e3b0c44298fc1c149afbf4c8996fb92427ae41e4649b934ca495991b7852b855
cmd.exe /c whoami
""".strip().split("\n")

with open("sample_syslog.txt", "w") as f:
    for line in sample_logs:
        f.write(line + "\n")
```

```python
import re

ioc_keywords = [
    r"\b45\.83\.120\.24\b",  # Example bad IP
    r"powershell\.exe",
    r"cmd\.exe",
    r"mimikatz",
    r"[a-f0-9]{64}"  # SHA-256 hash pattern
]
```

```python
def scan_logs_for_iocs(file_path):
    with open(file_path, "r") as f:
        lines = f.readlines()

    print("🔍 Scanning logs for Indicators of Compromise...\n")
    for line in lines:
        for pattern in ioc_keywords:
            if re.search(pattern, line, re.IGNORECASE):
                print(f"⚠️ IOC Match Found: {line.strip()}")
                break
```

```python
scan_logs_for_iocs("sample_syslog.txt")
```

#Malware Sandboxing (Simulated)

```python
import json

sample_report = {
    "target": {
        "file": {
            "name": "ransomware_sample.exe",
            "hashes": {
                "sha256": "9e107d9d372bb6826bd81d3542a419d6aeae6b5578cfb1767fa30d74c0edb037"
            }
        }
    },
    "network": {
        "hosts": [
            {"ip": "185.244.25.9"},
            {"ip": "92.222.139.45"}
        ],
        "dns": [
            {"request": "malicious-site.biz"},
            {"request": "stealer.net"}
        ]
    }
}

with open("malware_report.json", "w") as f:
    json.dump(sample_report, f, indent=4)

print("✅ Sample report written.")
```

```python
def parse_malware_report(report_path):
    with open(report_path, 'r') as f:
        data = json.load(f)

    print("📊 Malware Report Summary:")

    try:
        print(f"\n📁 File Name: {data.get('target', {}).get('file', {}).get('name')}")
        print(f"🔍 SHA256: {data.get('target', {}).get('file', {}).get('hashes', {}).get('sha256')}")
    except:
        print("❌ Failed to extract file info.")

    try:
        print("\n🌐 Contacted IPs:")
        for conn in data.get('network', {}).get('hosts', []):
            print(f" - {conn.get('ip')}")
    except:
        print("❌ No contacted IPs found.")

    try:
        print("\n🌐 Queried Domains:")
        for req in data.get('network', {}).get('dns', []):
            print(f" - {req.get('request')}")
    except:
        print("❌ No DNS queries found.")
```

```python
parse_malware_report("malware_report.json")
```

```python
from scapy.all import rdpcap

def analyze_pcap_scapy(file_path):
    print("🔍 Reading packets using Scapy...")
    packets = rdpcap(file_path)

    print(f"📦 Total packets: {len(packets)}\n")
    for i, pkt in enumerate(packets[:10]):
        print(f"Packet {i+1}:")
        pkt.show()
        print("-" * 40)

# Run this:
analyze_pcap_scapy("/content/4SICS-GeekLounge-151021.pcap")
```
